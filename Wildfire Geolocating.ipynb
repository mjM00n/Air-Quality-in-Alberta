```python
wildfiredf = pd.read_csv('Alberta_wildfires_cleaned.csv')
wildfiredf.head()
```

```python
key = "efd957fa4a304d098a3fe991f3eda5a6"
geocoder = OpenCageGeocode(key)

def municipality(lat, lon):
    try:
        result = geocoder.reverse_geocode(lat, lon)
        if result:
            components = result[0]['components']
            return components.get('county')
    except Exception as e:
        print(f"Error: {e}")
        return None


wildfiredf['Municipality'] = wildfiredf.apply(
    lambda row: municipality(row['fire_location_latitude'], row['fire_location_longitude']), axis=1)

print(wildfiredf[['fire_location_latitude', 'fire_location_longitude', 'Municipality']].head(10))

# https://www.youtube.com/watch?v=u-kkE4yA-z0&ab_channel=OpenCageGeocoding
# https://opencagedata.com/tutorials/geocode-in-python
```
