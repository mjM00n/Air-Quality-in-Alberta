# Wildfire Transformation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wildfiredf = pd.read_csv('Alberta_wildfires_cleaned_Geo.csv')
wildfiredf.head()
```

```python
fire_counts = wildfiredf.groupby(['fire_year', 'Municipality']).size().reset_index(name = 'total_fires')
fire_counts.to_csv("fire_counts.csv", index=False)
```

```python
fire_size_sum = wildfiredf.groupby(['fire_year', 'Municipality'])['current_size'].sum().reset_index(name='total_fire_size')
fire_size_sum.to_csv("fire_size_sum.csv", index=False)                                                                                    
```
